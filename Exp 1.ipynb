```python
from scipy.stats import zscore

numeric_cols = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
df[numeric_cols] = df[numeric_cols].astype(float)

z_scores = df[numeric_cols].apply(zscore, nan_policy='omit')


outliers = (z_scores.abs() > 3)

for col in numeric_cols:
    print(f"\nOutliers in {col}:")
    outlier_rows = df[outliers[col]]
    print(outlier_rows if not outlier_rows.empty else "No outliers found.")
```

```python
correlation_matrix = df[['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].corr()
```

```python
high_positive_corr = [(i, j, correlation_matrix.loc[i, j])
                      for i in correlation_matrix.columns
                      for j in correlation_matrix.columns
                      if i != j and correlation_matrix.loc[i, j] > 0.5]

high_negative_corr = [(i, j, correlation_matrix.loc[i, j])
                      for i in correlation_matrix.columns
                      for j in correlation_matrix.columns
                      if i != j and correlation_matrix.loc[i, j] < -0.5]

no_corr = [(i, j, correlation_matrix.loc[i, j])
           for i in correlation_matrix.columns
           for j in correlation_matrix.columns
           if i != j and abs(correlation_matrix.loc[i, j]) < 0.1]
```

```python
print("High Positive Correlations:")
for feature1, feature2, corr in high_positive_corr:
    print(f"{feature1} and {feature2}: {corr:.2f}")

print("\nHigh Negative Correlations:")
for feature1, feature2, corr in high_negative_corr:
    print(f"{feature1} and {feature2}: {corr:.2f}")

print("\nNo Correlation:")
for feature1, feature2, corr in no_corr:
    print(f"{feature1} and {feature2}: {corr:.2f}")
```

```python
import seaborn as sns
from scipy.stats import skew

numerical_features = df.select_dtypes(include=[np.number])

num_features = len(numerical_features.columns)
fig, axes = plt.subplots(nrows=(num_features // 3) + 1, ncols=3, figsize=(15,5))
axes = axes.flatten() 

for i, feature in enumerate(numerical_features.columns):
    sns.histplot(df[feature], kde=True, bins=30, ax=axes[i])
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel("Frequency")

for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

plt.subplots_adjust(hspace=0.5) 
plt.show()

skewness = numerical_features.apply(lambda x: skew(x.dropna()))

right_skewed = skewness[skewness > 0.5].index.tolist()
left_skewed = skewness[skewness < -0.5].index.tolist()
no_skew = skewness[(skewness >= -0.5) & (skewness <= 0.5)].index.tolist()

print("Right Skewed Features:", right_skewed)
print("Left Skewed Features:", left_skewed)
print("No Skew Features:", no_skew)
```

```python
# Select categorical columns
categorical_features = df.select_dtypes(include=['object'])

for feature in categorical_features.columns:
    plt.figure(figsize=(4,2))
    sns.countplot(data=df, x=feature)
    plt.title(f"Bar Plot of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

num_features = len(numerical_features.columns)

fig, axes = plt.subplots(nrows=2 * ((num_features // 3) + 1), ncols=3, figsize=(15,10 ))
axes = axes.flatten()

for i, feature in enumerate(numerical_features.columns):
    sns.swarmplot(data=df, x=feature, size=3, ax=axes[i])
    axes[i].set_title(f"Swarm Plot of {feature}")
    axes[i].set_xlabel(feature)

for j, feature in enumerate(numerical_features.columns):
    sns.violinplot(data=df, x=feature, ax=axes[num_features + j])
    axes[num_features + j].set_title(f"Violin Plot of {feature}")
    axes[num_features + j].set_xlabel(feature)

for k in range(num_features * 2, len(axes)):
    fig.delaxes(axes[k])

plt.subplots_adjust(hspace=0.5, wspace=0.4)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(4,2))
sns.scatterplot(data=df, x=numerical_features.columns[1], y=numerical_features.columns[2])
plt.title(f"Scatter Plot Between {numerical_features.columns[1]} and {numerical_features.columns[2]}")
plt.xlabel(numerical_features.columns[1])
plt.ylabel(numerical_features.columns[2])
plt.show()
```
